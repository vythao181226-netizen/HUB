```python
# ============================================================
# HỆ THỐNG DỰ BÁO NỢ XẤU NGÂN HÀNG
# ============================================================

from collections import deque

# ============================================================
# 1. CẤU TRÚC DỮ LIỆU
# ============================================================
khachhang_queue = deque()
lich_su_rui_ro_stack = []
khachhang_hash_table = {}

# ============================================================
# 2. NHẬP DỮ LIỆU KHÁCH HÀNG
# ============================================================
def dulieu_khachhang ():
    khachhang_queue.clear()
    khachhang_hash_table.clear()

    n = int(input("Nhập số lượng khách hàng: "))

    for i in range(n):
        print(f"\n--- Khách hàng {i + 1} ---")
        khach_hang = {
            "id": i + 1,
            "ten": input("Tên khách hàng: "),
            "ngay_qua_han": int(input("Số ngày quá hạn: ")),
            "du_no": int(input("Dư nợ (VND): ")),
            "thu_nhap": int(input("Thu nhập hàng tháng (VND): ")),
            "so_lan_tre_han": int(input("Số lần trễ hạn trước đây: "))
        }

        khachhang_queue.append(khach_hang)

        khachhang_hash_table[khach_hang["id"]] = {
            "info": khach_hang,
            "history": []
        }

# ============================================================
# 3. HÀM CHẤM ĐIỂM
# ============================================================
def diem_qua_han(ngay_qua_han):
    if ngay_qua_han > 120: return 60
    if ngay_qua_han > 90: return 50
    if ngay_qua_han > 30: return 30
    return 10

def diem_du_no(du_no):
    if du_no > 500_000_000: return 40
    if du_no > 300_000_000: return 30
    if du_no >= 100_000_000: return 20
    return 10

def diem_thu_nhap(thu_nhap):
    if thu_nhap < 10_000_000: return 30
    if thu_nhap <= 20_000_000: return 20
    return 10

def diem_so_lan_tre_han(so_lan_tre_han):
    if so_lan_tre_han >= 5: return 40
    if so_lan_tre_han >= 3: return 30
    if so_lan_tre_han > 0: return 20
    return 5

# ============================================================
# 4. TÍNH ĐIỂM TỔNG
# ============================================================
def diem_tong(khach_hang):
    diem = 0
    diem += diem_qua_han(khach_hang["ngay_qua_han"])
    diem += diem_du_no(khach_hang["du_no"])
    diem += diem_thu_nhap(khach_hang["thu_nhap"])
    diem += diem_so_lan_tre_han(khach_hang["so_lan_tre_han"])
    return diem

# ============================================================
# 5. PHÂN LOẠI & CẢNH BÁO
# ============================================================
def muc_do_rui_ro(diem):
    if diem >= 130: return "Rất cao"
    if diem >= 100: return "Cao"
    if diem >= 60:  return "Trung bình"
    return "Thấp"

def canh_bao_som(diem):
    if diem >= 100: return "⚠ CẢNH BÁO NỢ XẤU"
    if diem >= 70:  return "⚠ THEO DÕI"
    return "Bình thường"

# ============================================================
# 6. ĐÁNH GIÁ THEO KỲ
# ============================================================
def danh_gia_theo_ky(ten_ky):
    print(f"\n=== ĐÁNH GIÁ KỲ: {ten_ky} ===\n")

    while khachhang_queue:
        khach_hang = khachhang_queue.popleft()
        diem = diem_tong(khach_hang)
        rui_ro = muc_do_rui_ro(diem)
        canh_bao = canh_bao_som(diem)

        ket_qua = {
            "ky": ten_ky,
            "diem": diem,
            "rui_ro": rui_ro,
            "canh_bao": canh_bao
        }

        lich_su_rui_ro_stack.append({
            "id": khach_hang["id"],
            "ten": khach_hang["ten"],
            **ket_qua
        })

        khachhang_hash_table[khach_hang["id"]]["history"].append(ket_qua)

        print(f"KH: {khach_hang['ten']}")
        print(f"  Điểm: {diem}")
        print(f"  Rủi ro: {rui_ro}")
        print(f"  Trạng thái: {canh_bao}")
        print("-" * 50)

# ============================================================
# 7. TRA CỨU KHÁCH HÀNG BẰNG HASH TABLE
# ============================================================
def tra_cuu_khach_hang():
    cid = int(input("Nhập ID khách hàng cần tra cứu: "))

    if cid not in khachhang_hash_table:
        print("Không tồn tại khách hàng!")
        return

    du_lieu_khach_hang = khachhang_hash_table[cid]
    print("\n--- THÔNG TIN KHÁCH HÀNG ---")
    print(du_lieu_khach_hang["info"])

    print("\n--- LỊCH SỬ RỦI RO ---")
    for h in du_lieu_khach_hang["history"]:
        print(
            f"Kỳ: {h['ky']} | "
            f"Điểm: {h['diem']} | "
            f"Rủi ro: {h['rui_ro']} | "
            f"Trạng thái: {h['canh_bao']}"
        )

# ============================================================
# 8. THỐNG KÊ TOÀN HỆ THỐNG
# ============================================================
def bao_cao_thong_ke():
    thong_ke = {"Rất cao": 0, "Cao": 0, "Trung bình": 0, "Thấp": 0}

    for r in lich_su_rui_ro_stack:
        thong_ke[r["rui_ro"]] += 1

    print("\n=== THỐNG KÊ DANH MỤC ===")
    for k, v in thong_ke.items():
        print(f"{k}: {v}")

# ============================================================
# 9. XEM LỊCH SỬ ĐÁNH GIÁ (STACK)
# ============================================================
def lich_su_danh_gia():
    print("\n=== LỊCH SỬ ĐÁNH GIÁ (MỚI NHẤT TRƯỚC) ===")
    temp = lich_su_rui_ro_stack.copy()

    while temp:
        r = temp.pop()
        print(
            f"Kỳ: {r['ky']} | "
            f"KH: {r['ten']} | "
            f"Điểm: {r['diem']} | "
            f"Rủi ro: {r['rui_ro']}"
        )

# ============================================================
# 10. MENU HỆ THỐNG
# ============================================================
def menu():
    while True:
        print("\n====== MENU ======")
        print("1. Nhập dữ liệu khách hàng")
        print("2. Đánh giá rủi ro theo kỳ")
        print("3. Tra cứu khách hàng (Hash Table)")
        print("4. Thống kê danh mục")
        print("5. Xem lịch sử đánh giá")
        print("0. Thoát")

        lua_chon = input("Chọn: ")

        if lua_chon == "1":
            dulieu_khachhang()
        elif lua_chon == "2":
            ky = input("Nhập tên kỳ (VD: Tháng 1): ")
            danh_gia_theo_ky(ky)
        elif lua_chon == "3":
            tra_cuu_khach_hang()
        elif lua_chon == "4":
            bao_cao_thong_ke()
        elif lua_chon == "5":
            lich_su_danh_gia()
        elif lua_chon == "0":
            print("=== KẾT THÚC HỆ THỐNG ===")
            break
        else:
            print("Lựa chọn không hợp lệ!")

# ============================================================
# 11. CHẠY CHƯƠNG TRÌNH
# ============================================================
menu()
```
