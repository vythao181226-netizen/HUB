```python
class SimpleHashTable:
    def __init__(self, size):
        self.size = size
        self.table = [None] * size

    def hash_function(self, key):
        # Giả sử Key là Số tài khoản (Số nguyên)
        # Hàm băm trả về vị trí Index bằng phép chia lấy dư
        return key % self.size

    def insert(self, key, value):
        index = self.hash_function(key)
        self.table[index] = value
        print(f"Dữ liệu {value} được lưu tại Index: {index}")

# Khởi tạo bảng băm có 100 vị trí lưu trữ
bank_storage = SimpleHashTable(100)
bank_storage.insert(10925, "Giao dịch 5.600.000 VND") # Index 25
```

```python
import hashlib

# Hash Table lưu trữ user (Giả lập Database)
users = {}

def hash_password(password):
    return hashlib.sha256(password.encode()).hexdigest()

# Lưu mật khẩu đã băm
users["user01"] = hash_password("123456")

# Kiểm tra mật khẩu
print(users["user01"] == hash_password("123456"))
```

```python
import hashlib

# Hàm băm mã tài khoản
def hash_account(account_number):
    return hashlib.sha256(account_number.encode()).hexdigest()

# Hệ thống sàng lọc giao dịch
def screening_system(transaction_account, blacklist):
    # Băm mã tài khoản
    hashed_account = hash_account(transaction_account)

    # Tra cứu trong Hash Table
    if hashed_account in blacklist:
        return "Giao dịch bị chặn"
    else:
        return "Giao dịch được duyệt"
```

```python
blacklist = {
    hash_account("123456789"),
    hash_account("987654321"),
}

# Kiểm tra giao dịch
transaction_account = "123456789"
result = screening_system(transaction_account, blacklist)
print(result)  # Kết quả: Giao dịch bị chặn
```

```python
import hashlib

# 1. KHỞI TẠO DỮ LIỆU (Database)
transaction_db = {
    "GD001": "Chuyển tiền 5.000.000 VND",
    "GD002": "Thanh toán tiền điện",
    "GD003": "Rút tiền ATM"
}
# Blacklist chứa mã băm của tội phạm (VD: tài khoản '123456789')
blacklist_db = { hashlib.sha256("123456789".encode()).hexdigest() }

# 2. CÁC CHỨC NĂNG
def tra_cuu_nhanh():
    ma = input("Nhập mã GD: ")
    # Tra cứu O(1)
    print(f"-> Kết quả: {transaction_db.get(ma, 'Không tìm thấy')}")

def kiem_tra_aml():
    tk = input("Nhập số TK cần check: ")
    tk_hash = hashlib.sha256(tk.encode()).hexdigest()

    # Kiểm tra O(1) trong Blacklists
    if tk_hash in blacklist_db:
        print("CẢNH BÁO: Tài khoản nằm trong Blacklist -> CHẶN!")
    else:
        print("Tài khoản hợp lệ -> DUYỆT.")

# 3. CHƯƠNG TRÌNH CHÍNH
if __name__ == "__main__":
    while True:
        print("\n--- CORE BANKING SYSTEM ---")
        chon = input("1. Tra cứu GD | 2. Check AML | 3. Thoát: ")

        if chon == '1': tra_cuu_nhanh()
        elif chon == '2': kiem_tra_aml()
        elif chon == '3': break
```
