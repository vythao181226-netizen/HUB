```python
import pdfplumber
import pandas as pd
import requests
def collect_table_from_drive(url, page_num):
    try:
        file_id = url.split('/')[-2]
        direct_link = f'https://drive.google.com/uc?id={file_id}'
        local_path = "chude1.pdf"

        response = requests.get(direct_link)
        with open(local_path, 'wb') as f:
            f.write(response.content)

        with pdfplumber.open(local_path) as pdf:
            page = pdf.pages[page_num - 1]
            table = page.extract_table()

            if table:
                df = pd.DataFrame(table[1:], columns=table[0])
                print(f"--- Đã thu thập dữ liệu trang {page_num} thành công! ---")
                return df
        return None
    except Exception as e:
        print(f"Lỗi quy trình 1-2: {e}")
        return None
drive_url = 'https://drive.google.com/file/d/1ryF9_RvO8yL7MFIt7F66qYsMBWY0kxeB/view?usp=sharing'
```

```python
def extract_complete_model_details(path):
    try:
        import pdfplumber
        with pdfplumber.open(path) as pdf:
            full_combined_text = ""
            for i in [3, 4]:
                full_combined_text += pdf.pages[i].extract_text() + "\n"

            print("--- MÔ HÌNH NGHIÊN CỨU ---")
            print("Phương trình: Y1 = α0 + α1X1 + α2X2 + α3X3 + α4X4 + α5X5 + α6X6 + α7X7 + α8X8 + Ei")
            print("\nTrong đó:")

            lines = full_combined_text.split('\n')

            target_codes = ["Y1:", "X1:", "X2:", "X3:", "X4:", "X5:", "X6:", "X7:", "X8:", "Ei:"]

            history = set()

            for line in lines:
                for code in target_codes:
                    if code in line:
                        start_pos = line.find(code)
                        content = line[start_pos:].strip()

                        if content not in history:
                            print(content)
                            history.add(content)
                        break

    except Exception as e:
        print(f"Lỗi khi trích xuất: {e}")
extract_complete_model_details("chude1.pdf")
```

```python
import pdfplumber
import pandas as pd
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)

def extract_pearson_table(path):
    try:
        with pdfplumber.open(path) as pdf:
            page = pdf.pages[4]

            table = page.extract_table(table_settings={
                "vertical_strategy": "lines",
                "horizontal_strategy": "lines",
            })

            if not table:
                print("Không tìm thấy bảng.")
                return

            df = pd.DataFrame(table)

            df.columns = df.iloc[0]
            df = df.drop(0)
            df = df.map(lambda x: str(x).replace('\n', ' ') if x else "")
            df.columns.values[0] = "Variable"
            df.columns.values[1] = "Measure"

            print("\n--- BẢNG 2: THỐNG KÊ MÔ TẢ MẪU NGHIÊN CỨU ---")
            print(df.to_markdown(index=False))

    except Exception as e:
        print(f"Lỗi: {e}")

extract_pearson_table("chude1.pdf")
```

```python
def extract_vif_data_correctly(path):
    try:
        import pdfplumber
        import re

        with pdfplumber.open(path) as pdf:
            page = pdf.pages[5]
            text = page.extract_text()

            print("--- BẢNG 3: KIỂM ĐỊNH ĐA CỘNG TUYẾN ---")
            print(f"{'Biến':<10} | {'Tolerance':<15} | {'VIF':<10}")
            print("-" * 40)

            lines = text.split('\n')
            target_vars = ["SIZE", "ROA", "ROS", "ROE", "DR", "AS", "WCR", "GDP"]

            for line in lines:
                for var in target_vars:
                    if var in line:
                        numbers = re.findall(r'\d+[\.,]\d+', line)

                        if len(numbers) >= 2:
                            tolerance = numbers[-2]
                            vif = numbers[-1]
                            print(f"{var:<10} | {tolerance:<15} | {vif:<10}")
                        break

    except Exception as e:
        print(f"Lỗi khi trích xuất số liệu: {e}")

extract_vif_data_correctly("chude1.pdf")
```

```python
def extract_autocorrelation_table(path):
    try:
        import pdfplumber
        import re

        with pdfplumber.open(path) as pdf:
            if len(pdf.pages) <= 4:
                print("PDF không đủ trang.")
                return

            page = pdf.pages[4]
            text = page.extract_text()

            if not text:
                print("Không đọc được nội dung trang.")
                return

            print("--- BẢNG 4: KẾT QUẢ KIỂM ĐỊNH TỰ TƯƠNG QUAN ---")
            print(f"{'Model':<8} | {'R':<10} | {'R Square':<10} | {'Adj R2':<10} | {'Std Error':<12} | {'Durbin-Watson':<15}")
            print("-" * 80)

            lines = text.split('\n')

            for line in lines:
                if re.search(r'\b1\b', line):

                    numbers = re.findall(r'\d+[\.,]\d+', line)

                    if len(numbers) >= 5:
                        r = numbers[0]
                        r_square = numbers[1]
                        adj_r2 = numbers[2]
                        std_error = numbers[3]
                        dw = numbers[4]

                        print(f"{'1':<8} | {r:<10} | {r_square:<10} | {adj_r2:<10} | {std_error:<12} | {dw:<15}")
                        break

    except Exception as e:
        print(f"Lỗi khi trích xuất: {e}")

extract_autocorrelation_table("chude1.pdf")
```

```python
import pdfplumber
import pandas as pd
def extract_correct_pearson_table(path, page_index):
    try:
        with pdfplumber.open(path) as pdf:
            page = pdf.pages[page_index]

            tables = page.extract_tables({
                "vertical_strategy": "lines",
                "horizontal_strategy": "lines"
            })

            if not tables:
                print("Không tìm thấy bảng.")
                return None
            target_table = None
            for t in tables:
                text = str(t).lower()
                if "pearson" in text or "sig. (2-tailed)" in text:
                    target_table = t
                    break

            if not target_table:
                target_table = tables[-1]
            raw_df = pd.DataFrame(target_table)

            raw_df = raw_df.applymap(
                lambda x: str(x).replace('\n', ' ').strip() if x else ""
            )

            headers = raw_df.iloc[0].tolist()
            df = raw_df[1:].reset_index(drop=True)

            headers = [
                h if h != "" else f"col_{i}"
                for i, h in enumerate(headers)
            ]

            df.columns = headers
            df.rename(columns={
                df.columns[0]: "Variable",
                df.columns[1]: "Measure"
            }, inplace=True)

            def find_previous_var(idx):
                for j in range(idx, -1, -1):
                    v = str(df.iloc[j]["Variable"]).strip()
                    if v != "":
                        return v
                return None

            clean_rows = []
            i = 0

            while i < len(df):
                row = df.iloc[i]

                var = str(row["Variable"]).strip()
                measure = str(row["Measure"]).strip().lower()

                if "pearson" in measure:

                    current_var = var if var != "" else find_previous_var(i)

                    data = row[2:].tolist()

                    if i + 1 < len(df):
                        next_row = df.iloc[i + 1]
                        next_measure = str(next_row["Measure"]).lower()

                        if "correlation" in next_measure or all(x == "" for x in data):
                            data = next_row[2:].tolist()
                            i += 1

                    clean_rows.append([current_var, "Pearson Correlation"] + data)

                elif "sig" in measure:
                    current_var = var if var != "" else find_previous_var(i)
                    clean_rows.append([current_var, "Sig. (2-tailed)"] + row[2:].tolist())

                elif measure == "n":
                    current_var = var if var != "" else find_previous_var(i)
                    clean_rows.append([current_var, "N"] + row[2:].tolist())

                i += 1

            final_df = pd.DataFrame(clean_rows, columns=df.columns)

            def convert_num(x):
                try:
                    return float(
                        str(x).replace(',', '.').replace('*', '')
                    )
                except:
                    return x

            final_df = final_df.applymap(convert_num)
            final_df["Variable"] = final_df["Variable"].replace([None, "", "None"], "CR")

            print("\n--- BẢNG 5: PHÂN TÍCH TƯƠNG QUAN PEARSON ---")
            print(final_df.to_markdown(index=False))
            final_df["Variable"] = final_df["Variable"].replace([None, "", "None"], "CR")
            return final_df

    except Exception as e:
        print(f"Lỗi: {e}")

extract_correct_pearson_table("chude1.pdf", 5)
```

```python
import pdfplumber
import pandas as pd
def extract_model_summary(path, page_index=5):
    try:
        with pdfplumber.open(path) as pdf:
            page = pdf.pages[5]

            table = page.extract_table({
                "vertical_strategy": "lines",
                "horizontal_strategy": "lines"
            })

            if not table:
                print(f"Không tìm thấy bảng ở trang {page_index}.")
                return None

            df = pd.DataFrame(table)
            header_top = df.iloc[0].fillna("").tolist()
            header_bottom = df.iloc[1].fillna("").tolist()

            final_headers = []
            current_top = ""

            for top, bottom in zip(header_top, header_bottom):
                top = top.replace('\n', ' ').strip()
                bottom = bottom.replace('\n', ' ').strip()

                if top != "":
                    current_top = top

                if current_top and bottom and current_top != bottom:
                    final_headers.append(f"{current_top}_{bottom}")
                else:
                    final_headers.append(bottom if bottom else top)

            df.columns = final_headers
            df = df.drop([0, 1]).reset_index(drop=True)

            df = df[df.iloc[:, 0] != "a. Predictors: (Constant), GDP, AS, ROE, ROS, SIZE, ROA, DR, WCR"]

            df = df.map(lambda x: str(x).replace('\n', ' ').strip() if x else "")

            print("\n--- BẢNG 7: KIỂM ĐỊNH SỰ PHÙ HỢP CỦA MÔ HÌNH ---")
            print(df.to_markdown(index=False))

            return df

    except Exception as e:
        print(f"Lỗi: {e}")

extract_model_summary("chude1.pdf", page_index=5)
```
